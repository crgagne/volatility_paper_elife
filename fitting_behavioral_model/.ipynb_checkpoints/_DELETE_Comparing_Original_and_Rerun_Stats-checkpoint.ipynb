### Posterior Stats

```python
def print_posteriors(params=None,
                     trace_dev=None,
                     which='lr',
                     group_param='u',
                     name=None):
    if name is None:
        if group_param=='u':
            name='group'
        else:
            name='trait'
    print('posterior probability for '+name+' effects')
    print()
    for pi,param in enumerate(params):
        if which in param:
            P_greater = np.mean(trace_dev[group_param][:,pi][:]>0)
            interval = pm.stats.hpd(trace_dev[group_param][:,pi].flatten(),alpha=0.05)
            if P_greater<0.5:
                P_greater=1-P_greater
            if np.sign(interval[0])==np.sign(interval[1]):#P_greater>0.95:
                bold1='\033[1m'
                bold2='\033[0m'+'*'
            else:
                bold1=''
                bold2=''
            
            print(param+': \t P(theta<>0)='+bold1+str(np.round(P_greater,3))+bold2+' \t 95%-CI ['+str(np.round(interval[0],3))+','+str(np.round(interval[1],3))+']')
```

```python
for param in ['lr','Binv','Amix','Bc','mag','lr_c']:
    for i,gp in enumerate(['u','u_PC1','u_PC2','u_PC3',]):
        print('orig')
        print_posteriors(params=model_orig.params,
                             which=param,
                             trace_dev=trace_orig,
                             group_param=gp,name=gp)
        print('')
        print('new')
        print_posteriors(params=model.params,
                             which=param,
                             trace_dev=trace,
                             group_param=gp,name=gp)
        print('')
        print('')
```
