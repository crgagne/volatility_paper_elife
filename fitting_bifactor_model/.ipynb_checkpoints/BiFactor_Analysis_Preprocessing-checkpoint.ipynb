# Mscl Functions

```python
def convert_diagnosis_to_color(diag):
    color=[]
    for d in diag:
        if d=='control_mfmri':
            color.append(sns.color_palette()[0])
        elif d=='control_cdm':
            color.append(sns.color_palette()[1])
        elif d=='MDD':
            color.append(sns.color_palette()[2])
        elif d=='GAD':
            color.append(sns.color_palette()[3])
        elif d=='amt':
            color.append(sns.color_palette()[4])
    return(color)
```

## Collapse Response categories with <2% 
in any (because it'll be hard to estimate polychoric if any are like that 

```python
Q_items_confirmatory_col = Q_items_confirmatory.copy()
Q_items_cdm_col = Q_items_cdm.copy()
Q_items_online_col = Q_items_online.copy()
```

```python
Q_items_confirmatory_col.shape
```

```python
value_less_than1=0
for item in range(128):
    plt.figure(figsize=(12,4))
    
    # plot janines
    ax1=plt.subplot(1,3,1)
    x1 =Q_items_confirmatory_col.iloc[item,:].values
    plt.hist(x1,alpha=0.5)

    # plot cdm 
    ax2=plt.subplot(1,3,2)
    x2 =Q_items_cdm_col.iloc[item,:].values
    plt.hist(x2,alpha=0.5)
    
    # plot online 
    ax3=plt.subplot(1,3,3)
    x3 =Q_items_online_col.iloc[item,:].values
    
    missing = np.all(np.isnan(x3))
    
    if not missing: # missing scales 
        plt.hist(x3,alpha=0.5)
    
    # calculate value counts on janines (percent for each response)
    vcounts_j = Q_items_confirmatory_col.iloc[item,:].value_counts()
    vcounts_less_j = (vcounts_j/len(x1)<0.02)
    tmp = vcounts_less_j.index[vcounts_less_j==True]
#     if len(tmp)>0:
#         print('j')
#         print(tmp[:])

    vcounts_c = Q_items_cdm_col.iloc[item,:].value_counts()
    vcounts_less_c = (vcounts_c/len(x2)<0.02)
    tmp = vcounts_less_c.index[vcounts_less_c==True]
#     if len(tmp)>0:
#         print('c')
#         print(tmp[:])

    if not missing:
        vcounts_o = Q_items_online_col.iloc[item,:].value_counts()
        vcounts_less_o = (vcounts_o/len(x3)<0.02)
        tmp = vcounts_less_o.index[vcounts_less_o==True]
#     if len(tmp)>0:
#         print('o')
#         print(tmp[:])
#     print('')
    
    if not missing: 
        vcounts_less = vcounts_less_j|vcounts_less_c|vcounts_less_o
    else:
        vcounts_less = vcounts_less_j|vcounts_less_c
        
    # adjust for all 3 datasets 
    for i,x in enumerate([x1,x2,x3]):
        
        # adjust 1 bin
        if len(vcounts_less[vcounts_less])==1:
            if item not in [0,18,100]: # usual  
                value_less_than1 = list(vcounts_less[vcounts_less].index)[0]
                x[x==value_less_than1]=value_less_than1-1
            elif item in [100]:
                value_less_than1 = list(vcounts_less[vcounts_less].index)[0]+1 # drop 3 into 2
                x[x==value_less_than1]=value_less_than1-1
            else:
                value_less_than1 = list(vcounts_less[vcounts_less].index)[0]
                x[x==value_less_than1]=value_less_than1+1
                
        # adjust 2 bins 
        elif len(vcounts_less[vcounts_less])==2:
            values_less_than1 = list(vcounts_less[vcounts_less].index)
            values_less_than1.sort()
            x[x==values_less_than1[-1]]=values_less_than1[-1]-2
            x[x==values_less_than1[-2]]=values_less_than1[-2]-1
        
        # adjust 3 bins 
        elif len(vcounts_less[vcounts_less])==3:
            values_less_than1 = list(vcounts_less[vcounts_less].index)
            values_less_than1.sort()
            x[x==values_less_than1[-1]]=values_less_than1[-1]-3
            x[x==values_less_than1[-2]]=values_less_than1[-2]-2
            x[x==values_less_than1[-3]]=values_less_than1[-2]-1
        
        else:
            value_less_than1 = ''
            
        ## fix a few manually # 
        if i==1 and item==97:
            x[x==3.0]=2.0
        if i==1 and item==77:
            x[x==5.0]=4.0       
        if i==2 and item==77:
            x[x==5.0]=4.0   

            
        if i==0:
            plt.sca(ax1)
            Q_items_confirmatory_col.iloc[item,:]=x
        elif i==1:
            plt.sca(ax2)
            Q_items_cdm_col.iloc[item,:]=x
        elif i==2:
            if not missing:
                plt.sca(ax3)
                Q_items_online_col.iloc[item,:]=x
                
        if not missing or i<2:
            plt.hist(x,alpha=0.3)
        plt.title(str(item)+Q_labels_confirmatory.iloc[item]['Questionnaire']+\
                  Q_labels_confirmatory.iloc[item]['Question']+'  '+str(value_less_than1))
        
    # checking all the same categories
    assert np.all(np.sort(np.unique(x1[~np.isnan(x1)]))==np.sort(np.unique(x2[~np.isnan(x2)])))
    if not missing:
        assert np.all(np.sort(np.unique(x1[~np.isnan(x1)]))==np.sort(np.unique(x3[~np.isnan(x3)])))
```

## Reverse Collapsed Items

```python
Q_items_confirmatory_col_rev = Q_items_confirmatory_col.copy()
print(Q_items_confirmatory_col_rev.shape)

Q_items_cdm_col_rev = Q_items_cdm_col.copy()
print(Q_items_cdm_col_rev.shape)

Q_items_online_col_rev = Q_items_online_col.copy()
print(Q_items_online_col_rev.shape)
```

```python
rev_idx = (Q_labels_confirmatory['Rev']==1).values

# same reversals
np.all(rev_idx==(Q_labels_cdm['Rev']==1).values)
np.all(rev_idx==(Q_labels_online['Rev']==1).values)

rev_idx.shape
```

```python
# ok all the max's match 
print(np.where((Q_items_confirmatory_col.max(axis=1)==Q_items_cdm_col.max(axis=1))==False))
#non_matching=np.where((Q_items_confirmatory_col.max(axis=1)==Q_items_online_col.max(axis=1))==False)[0]
#Q_labels_online.loc[non_matching]
```

```python
# ok all the mins match 
print(np.where((Q_items_confirmatory_col.min(axis=1)==Q_items_cdm_col.min(axis=1))==False))
#non_matching=np.where((Q_items_confirmatory_col.min(axis=1)==Q_items_online_col.min(axis=1))==False)[0]
#Q_labels_online.loc[non_matching]
```

```python
##### Recalculate min and max
Q_labels_confirmatory['Max']=Q_items_confirmatory_col.max(axis=1)
Q_labels_cdm['Max']=Q_items_confirmatory_col.max(axis=1)
Q_labels_online['Max']=Q_items_confirmatory_col.max(axis=1)

Q_labels_confirmatory['Min']=Q_items_confirmatory_col.min(axis=1)
Q_labels_cdm['Min']=Q_items_confirmatory_col.min(axis=1)
Q_labels_online['Min']=Q_items_confirmatory_col.min(axis=1)
```

```python
Min_j = np.repeat(Q_labels_confirmatory.Min.values[:,np.newaxis],Q_items_confirmatory_col_rev.values.shape[1],axis=1)
Max_j = np.repeat(Q_labels_confirmatory.Max.values[:,np.newaxis],Q_items_confirmatory_col_rev.values.shape[1],axis=1)

Min_c = np.repeat(Q_labels_cdm.Min.values[:,np.newaxis],Q_items_cdm_col_rev.values.shape[1],axis=1)
Max_c = np.repeat(Q_labels_cdm.Max.values[:,np.newaxis],Q_items_cdm_col_rev.values.shape[1],axis=1)

Min_o = np.repeat(Q_labels_online.Min.values[:,np.newaxis],Q_items_online_col_rev.values.shape[1],axis=1)
Max_o = np.repeat(Q_labels_online.Max.values[:,np.newaxis],Q_items_online_col_rev.values.shape[1],axis=1)

```

```python
# reverse
Q_items_confirmatory_col_rev.loc[rev_idx,:]=Max_j[rev_idx,:]-Q_items_confirmatory_col_rev.loc[rev_idx,:]+Min_j[rev_idx,:]
Q_items_cdm_col_rev.loc[rev_idx,:]=Max_c[rev_idx,:]-Q_items_cdm_col_rev.loc[rev_idx,:]+Min_c[rev_idx,:]
Q_items_online_col_rev.loc[rev_idx,:]=Max_o[rev_idx,:]-Q_items_online_col_rev.loc[rev_idx,:]+Min_o[rev_idx,:]
```
